```python
import pycountry_convert as pc
import pandas as pd
# Load the dataset
file_path = 'Time on the internet.csv'
data = pd.read_csv(file_path, encoding='ascii')

# Function to get continent name from country name
def country_to_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
        return country_continent_name
    except:
        return 'Unknown'  # For entities that are not countries or have different names

# Apply the function to the 'Entity' column to create a new 'Continent' column
data['Continent'] = data['Entity'].apply(lambda x: country_to_continent(x))

# Display the updated dataframe with the new 'Continent' column
print(data.head())
```

```python
# Filter the data for the year 2020
data_2020 = data[data['Year'] == 2020]

# Group by 'Entity' and sum the 'Number of Internet users'
entity_internet_users = data_2020.groupby('Entity')['Number of Internet users'].sum().reset_index()

# Sort the data by 'Number of Internet users' to get the top entities
entity_internet_users_sorted = entity_internet_users.sort_values(by='Number of Internet users', ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 8), facecolor='white')
plt.bar(entity_internet_users_sorted['Entity'], entity_internet_users_sorted['Number of Internet users'], color='g')
plt.title('Top 10 Entities by Number of Internet Users in 2020')
plt.xlabel('Entity')
plt.ylabel('Number of Internet Users')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

data = pd.read_csv('Time on the internet.csv', encoding='ascii')

# Filter the data for the years 1900 to 2000
filtered_data = data[(data['Year'] >= 1900) & (data['Year'] <= 2000)]

# Group by country and sum the number of internet users
grouped_data = filtered_data.groupby('Entity')['Number of Internet users'].sum().reset_index()

# Sort the data by the number of internet users in descending order and select the top 5
top_5_countries = grouped_data.sort_values(by='Number of Internet users', ascending=False).head(5)

# Display the top 5 countries
top_5_countries
```

```python
# Since the United States is the only specific country listed, we'll create a pie chart for it
# However, given the context, it might not be meaningful to create a pie chart for a single country
# Instead, let's create a pie chart for the top entities including regions and income groups

import matplotlib.pyplot as plt

# Data for the entities with the highest number of internet users from 1900 to 2000
entities = [ 'High-income countries', 'United States', 'Europe', 'Asia']
users = [ 980617065, 460237493, 291379109, 233532495]

# Create a pie chart
plt.figure(figsize=(10, 7), facecolor='white')
plt.pie(users, labels=entities, autopct='%1.1f%%', startangle=140)
plt.title('Top Entities by Number of Internet Users (1900-2000)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Save the pie chart as an HTML file with embedded base64 image
plt.savefig('top_entities_internet_users_1900_2000.png')
plt.show()
```
